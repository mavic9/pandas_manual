### Task 1

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.arange(1, 8)
print(x)
y = np.array([3.5, 3.8, 4.2, 4.5, 5, 5.5, 7])
print(y)

plt.plot(x, y, 'b-')
plt.xlabel('x', fontsize = 14)
plt.ylabel('y', fontsize = 14)
plt.show()
```

```python
plt.scatter(x, y)
plt.title('Scatter plot', fontsize = 16)
plt.xlabel('x', fontsize = 14)
plt.ylabel('y', fontsize = 14)
plt.show()
```

### Task 2

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 10, 51)
f = np.cos(t)

fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(t, f, 'g-')
ax.set_title('Figure 2', fontsize = 16)
ax.set_xlabel('t', fontsize = 14)
ax.set_ylabel('f = cos(t)', fontsize = 14)
ax.set_ylim(ymin=-2.5, ymax = 2.5)
ax.set_xlim([0.5, 9.5])
plt.show()
```

### Task 3

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 51)

y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)

fig, axs = plt.subplots(2, 2, figsize = (8, 6))
ax1, ax2, ax3, ax4 = axs.flatten()

# fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title('Figure y1', fontsize = 16)
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('Figure y2', fontsize = 16)

ax3.plot(x, y3)
ax3.set_title('Figure y3', fontsize = 16)

ax4.plot(x, y4)
ax4.set_title('Figure y4', fontsize = 16)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline


df = pd.read_csv('/home/viktor/Documents/GB_HW/numpy/datasets/creditcard.csv', sep=',')
df.head(10)
```

```python
class_data = df['Class'].value_counts()
print(class_data)
```

```python
plt.style.use('fivethirtyeight')

class_data.plot(kind = 'bar')
plt.title('Credit card operation class')
plt.show()

class_data.plot(kind = 'bar', logy = True)
plt.ylim(1, 300000)
plt.title('Credit card operation class')
plt.show()
```

```python

# filtering data on the variable "Class"
V1_normal = df.loc[(df['Class'] == 0), "V1"]
V1_frauds = df.loc[(df["Class"] == 1), "V1"]

# plot the histogram for a density of values in parameter V1 
plt.hist(V1_normal, bins = 20, ec='grey', color = 'gray', density = True, alpha = 0.5, label = 'Class 0')
plt.hist(V1_frauds, bins = 20, ec='grey', color = 'red', density = True, alpha = 0.5, label = 'Class 1')
plt.ylabel('Density of values')
plt.xlabel('Parameter V1')
plt.legend(loc='best')
plt.show()
```
